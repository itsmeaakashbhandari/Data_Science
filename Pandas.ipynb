```python
df.drop(3,inplace=True)
df.head()
```

```python
#to print labels
df.index
```

### How to deal with Na

```python
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
df.describe()
```

### First way to drop that whole row where NaN is present

```python
#how to deal with NaN
df.dropna() #it drop whol row
```

```python
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

## Second way is to fill them with the mean of that particular column 


```python
df.sw.fillna(df.sw.mean(), inplace=True)
df.pl.fillna(df.pl.mean(), inplace=True)
```

```python
df.head()
```

## Handling with String

```python
df['Gender']='Female'
df.iloc[0:10,6]='Male'

def f(s):
    if s=='Male':
        return 0
    else:
        return 1
df['Sex']=df.Gender.apply(f)
```
