# Data
1. Data is simply a collection or set of values in the form of text, words, numbers, pictures, audio or video.
2. It is most important asset for any organisation.

# How to store data?
1. Traditional way - on paper
2. File system
3. Databases

# Databases
1. it is an organized collection or repository of our data
2. it is a collection of interrelated data which helps in efficient retrieval, insertion and deletion of data.
3. users can perform different queries in order to perform actions based on their requirement.

# DBMS
1. Database management system.
2. it is an application which is used to maintain or manage databases.
3. Eg- Oracle, sybase, microsoft SQL server, PostgreSQL and more.

## Types of DBMS
1. Hierarchical DBMS
2. Network DBMS
3. Relatinal DBMS
4. Object oriented DBMS

# Relational Database
1.  it is a specific type of database that stores everything in relations or tables.
2. Tables have different row and columns.
3. row=tuple, column=attribute.
4. RDBMS is a software system which is used to maintain relational databases is termed as relational database management system(RDBMS).
# Primary key
1. It is a minimal set of columns or attributes that uniquely identifies row in a table.
2. main features-
   ##### 2.1 it must contain a unique value for each row of data.
   ##### 2.2 It cannot contain null values.
3. ex- StudentID in Student Table.   

# Foreign key
1. They are columns that point to or that matches the primary key columns in other tables.
2. The foreign key can be used to cross-reference tables. Foreign keys do not need to have unique values in the referencing relation.
 


# SQL- 
1. structured query language.
2. Language used to interact with relational database.
3. An SQL query is how you access the data. Using an SQL query, you can create and delete, or modify tables, as well as select, insert, and delete data from existing tables.
4. Ex of popular RDBMSs- SQLite, MySQL, PostgreSQL, Oracle DB, SQL Server.
