## Customise graph
### Color, marker, linewidth, labels foraxis, datasets, and the title, axis, grid, text.

```python
#plot function x^3
import numpy as np
x=np.arange(0,5,0.1) #(starting,end,difference)
y=x**3
plt.plot(x,y,color='black',marker='o')
plt.show()
```

```python
#plot function x^3
import numpy as np
x=np.arange(0,5,0.1) #(starting,end,difference)
y=x**3
plt.plot(x,y,color='black',linewidth=5)
plt.show()
```

### Label,title in graph

```python
#plot function x^3 and x^2
#color,marker,linewidth,labels for axis,datasets and the title
import numpy as np
x=np.arange(0,5,0.1) #(starting,end,difference)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth=2,label='x^3')
plt.plot(x,y2,color='red',linewidth=2,label='x^2')
plt.ylabel('y') #y axis label
plt.xlabel('x') #x axis label
plt.legend()  #to add label for lines
plt.title('Matplotlib demo') #title
plt.show()
```

```python
#to change axis values
import numpy as np
x=np.arange(0,5,0.1) #(starting,end,difference)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth=2,label='x^3')
plt.plot(x,y2,color='red',linewidth=2,label='x^2')

#axis 
plt.axis([0,10,0,100])  #([xStart,xEnd,yStart,yEnd])
plt.ylabel('y') #y axis label
plt.xlabel('x') #x axis label
plt.legend()  #to add label for lines
plt.title('Matplotlib demo') #title
plt.show()
```

```python
#to create a grid
import numpy as np
x=np.arange(0,5,0.1) 
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth=2,label='x^3')
plt.plot(x,y2,color='red',linewidth=2,label='x^2')
plt.ylabel('y') 
plt.xlabel('x') 

#grid
plt.grid()
plt.legend() 
plt.title('Matplotlib demo')
plt.show()
```

```python
#to add text
import numpy as np
x=np.arange(0,5,0.1) #(starting,end,difference)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth=2,label='x^3')
plt.plot(x,y2,color='red',linewidth=2,label='x^2')
plt.ylabel('y') #y axis label
plt.xlabel('x') #x axis label
plt.grid()

#add text
plt.text(2,80,'TEXT')  #(POSITION-X,POSITION-Y,'TEXT')
plt.legend()  #to add label for lines
plt.title('Matplotlib demo') #title
plt.show()
```

# Bubble Graph/Chart

```python
import matplotlib.pyplot as plt
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]


plt.scatter(year,salary, s=population, color='r') #(x,y,size)
plt.title('Bubble Chart')
plt.show()
```

```python
#opaque of color
plt.scatter(year,salary, s=population, color='r', alpha=0.5) #(x,y,size,opaque of color(lightness))
plt.title('Bubble Chart')
plt.show()
```

```python
import numpy as np
c=np.arange(len(year))
c
```

```python
#different color(c=c) by passing array
plt.scatter(year,salary, s=population, c=c, alpha=0.5) #(x,y,size,opaque of color(lightness))
plt.title('Bubble Chart')
plt.show()
```

```python
#bubble edge color
plt.scatter(year,salary, s=population, c=c, alpha=0.5, edgecolor='black')
plt.title('Bubble Chart')
plt.show()
```

```python
#change the marker
plt.scatter(year,salary, s=population, c=c, alpha=0.5,edgecolor='black',marker='^') #(x,y,size,opaque of color(lightness))
plt.title('Bubble Chart')
plt.show()
```

```python
#write text with marker
plt.scatter(year,salary, s=population, c=c, alpha=0.5,edgecolor='black',marker='^') #(x,y,size,opaque of color(lightness))
plt.title('Bubble Chart')
plt.text(year[0],salary[0]+0.5,population[0])
plt.text(year[1],salary[1]+0.5,population[1])
plt.show()
```

# Pie Graph/Chart

```python
sizes=[3,4,6,2]
plt.pie(sizes)
plt.axis('equal')
plt.show()
```

```python
#colors
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
plt.pie(sizes,colors=color)
plt.axis('equal')
plt.show()
```

```python
#labels
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
plt.pie(sizes,colors=color,labels=label)
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
#label inside pie chart
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
plt.pie(sizes,colors=color,labels=label,autopct='%.1f%%') #(label percentage- autopct=1 digit after decimal.%%-for percentage sign)
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
#highlighting particular class
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
#highlighting and taking little out
explode=[0.2,0,0,0]  #position for particular class that we want to highlight
plt.pie(sizes,colors=color,explode=explode,labels=label,autopct='%.1f%%') #(label percentage- autopct=1 digit after decimal.%%-for percentage sign)
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
#highlighting and taking little out
explode=[0.1,0.2,0,0]  #position for particular class that we want to highlight
plt.pie(sizes,colors=color,explode=explode,labels=label,autopct='%.1f%%') #(label percentage- autopct=1 digit after decimal.%%-for percentage sign)
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
#to make pie in a clockwise direction we use- counterclock=false
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
#highlighting and taking little out
explode=[0.1,0.2,0,0]  #position for particular class that we want to highlight
plt.pie(sizes,colors=color,explode=explode,labels=label,autopct='%.1f%%',counterclock=False)#clockwise
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
#to change the position of classes- startangle=100
label=['A','B','C','D']
sizes=[3,4,6,2]
color=['blue','purple','red','pink']
#highlighting and taking little out
explode=[0.1,0.2,0,0]  #position for particular class that we want to highlight
plt.pie(sizes,colors=color,explode=explode,labels=label,autopct='%.1f%%',counterclock=False,startangle=180)
plt.title('Split among classes ')
plt.axis('equal')
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='black')
plt.show()
```

```python
#change axis
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='black')
plt.axis([0,25,0,10])
plt.show()
```

# Bar Graph

```python
#different year
year=['year - 2012','year - 2013','year - 2014','year - 2015','year - 2016','year - 2017']
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.bar(year,population,width=0.6)
plt.show()

#here axis label are over laping
```

# How to decide Graph type?

## Bar Graphs- 
Where there are some few categories to compare. They are really good to look trends.
## Line Graphs-
Where ther are huge ampunt of categories to compare. And it is also really good to look trends. (ex- stock-price trends).
## Scatter graph-
Correlation between two variables.
## Pie Chart- 
Who gets what ,%,which gets how(ex- different sector budget of a country)
## Bubble chart
When we want to show two variables,  but ther is a third variable which we think more important when somebody visualize these two variable (year,salary.population)
## Histograms- 
Good to show frequency distrubution, data utilized in a particular range.
