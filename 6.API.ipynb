 # HTTP -
 ### Hyper text transfer protocol
 1. protocol is nothing but a set of rules which defines how to format and transmit the request to server and how to receive response from server.
 2. communication between client and server on internet happens using HTTP.
 3. Information is exchanged in the form of hypertest documents.
 4. A web browser may be the client, and an application on a computer that hosts a web site may be the server.
 
### Methods of HTTP


1. GET - Kind of read method, by this we can request any server/database for information.

2. POST - Kind of write method, we can update our database.

# JSON File
1. JSON - javascript object notation
2. most common formats for information retrieved, in API are JSON and XML.
3. Can be editted using text editor
4. it has two parts: Key and Value.

## JSON datatypes
1. primitive data types: number, string, boolean, null.
2. complex data types : array, object.

## JSON to Python


```python
#Object to dictionary
import json
json_data='{"Student":"Mohit"}'
python_data=json.loads(json_data)
print(python_data)
print(type(python_data))
print(python_data['Student'])
```

```python
#Array to list
json_data= '{"Fruits":["Apple","orange","mango","guava"]}'
python_data=json.loads(json_data)
print(type(python_data))
print(type(python_data["Fruits"]))  #here array converted into list

#here Object: Fruits (key) is converted into dictionary and Array (value) is converted into list.
print(len(python_data["Fruits"]))
for i in python_data["Fruits"]:
    print(i)
```
